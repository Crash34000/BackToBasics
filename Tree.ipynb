# Tree

## DFS


```python
adj={}
adj[0]=[1,2,3]
adj[1]=[4,5]
adj[2]=[]
adj[3]=[6]
adj[4]=[]
adj[5]=[7]
adj[6]=[]
adj[7]=[]


def dfs(s,e):
    # noeud s , arete e
    print(s+1)
    for u in adj[s]:
        if u!=e:
            dfs(u,s)
dfs(0,0)
```

## Compter nombre de noeud dans le subtree

DP ;)

```python
count=[0]*8
def dfs(s,e):
    count[s]=1
    # noeud s , arete e
    for u in adj[s]:
        if u!=e:
            dfs(u,s)
            count[s]+=count[u]
dfs(0,0)
print(count)
```
